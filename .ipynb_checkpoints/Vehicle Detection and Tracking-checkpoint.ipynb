## Vehicle Detection and Tracking
---
The goal is to create a pipeline to identify and treck vehicles in a video from a front-facing camera on a car.

I used a traditional Computer Vision approach for object detection: processing stages, feature extraction, spatial sampling and classification. Different algorithms and techniques can be applied for these stages, and the parameters for each stage are usually tuned by hand.

In comparison, a Deep Neural Network designed for object detection can perform these tasks using a complex interconnected architecture where the stages are not as distinct. Lower layers of a neural network, i.e. the ones closer to the input data, typically perform some equivalent of feature extraction, while higher layers can localize and classify simultaneously to produce detections.

```python
# Import all necessary libraries

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
import glob
import time
from sklearn.svm import LinearSVC, SVC
from sklearn.preprocessing import StandardScaler
from skimage.feature import hog
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from scipy.ndimage.measurements import label
import pandas as pd
```

### Helper functions for image processing and classification

```python
def get_hog_features(img, orient, pix_per_cell, cell_per_block, 
                        vis=False, feature_vec=True):
    # Call with two outputs if vis==True
    if vis == True:
        features, hog_image = hog(img, orientations=orient, 
                                  pixels_per_cell=(pix_per_cell, pix_per_cell),
                                  cells_per_block=(cell_per_block, cell_per_block), 
                                  transform_sqrt=False, 
                                  visualise=vis, feature_vector=feature_vec)
        return features, hog_image
    # Otherwise call with one output
    else:      
        features = hog(img, orientations=orient, 
                       pixels_per_cell=(pix_per_cell, pix_per_cell),
                       cells_per_block=(cell_per_block, cell_per_block), 
                       transform_sqrt=False, 
                       visualise=vis, feature_vector=feature_vec)
        return features

def bin_spatial(img, size=(32, 32)):
    color1 = cv2.resize(img[:,:,0], size).ravel()
    color2 = cv2.resize(img[:,:,1], size).ravel()
    color3 = cv2.resize(img[:,:,2], size).ravel()
    return np.hstack((color1, color2, color3))
                        
def color_hist(img, nbins=32):    #bins_range=(0, 256)
    # Histogram of the color channels separately
    channel1_hist = np.histogram(img[:,:,0], bins=nbins)
    channel2_hist = np.histogram(img[:,:,1], bins=nbins)
    channel3_hist = np.histogram(img[:,:,2], bins=nbins)
    # Concatenate the histograms into a single feature vector
    hist_features = np.concatenate((channel1_hist[0], channel2_hist[0], channel3_hist[0]))
    # Return the individual histograms, bin_centers and feature vector
    return hist_features

def draw_boxes(img, bboxes, color=(0, 0, 255), thick=6):
    # A function takes an image, a list of bounding boxes, 
    # and optional color tuple and line thickness as inputs
    # then draws boxes in that color on the output
    
    # Make a copy of the image
    draw_img = np.copy(img)
    # Iterate through the bounding boxes
    for bbox in bboxes:
        # Draw a rectangle given bbox coordinates
        cv2.rectangle(draw_img, bbox[0], bbox[1], color, thick)
    # Return the image copy with boxes drawn
    return draw_img

def SVM_result(X, y, model_output = False):
    # Fit a per-column scaler
    X_scaler = StandardScaler().fit(X)
    # Apply the scaler to X
    scaled_X = X_scaler.transform(X)
    # Split up data into randomized training and test sets
    rand_state = 42 #np.random.randint(0, 100)
    X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.2, random_state=rand_state)
    # Use a linear SVC 
    svc = LinearSVC()
    svc.fit(X_train, y_train)
    if model_output:
        return round(svc.score(X_test, y_test), 4), svc, X_scaler
    else:
        return round(svc.score(X_test, y_test), 4)
```

### Upload images into memory

```python
# Image paths
pathNonVehicles1 = 'data/non-vehicles/Extras/*.png'
pathNonVehicles2 = 'data/non-vehicles/GTI/*.png'

pathVehicles1 = 'data/vehicles/GTI_Far/*.png'
pathVehicles2 = 'data/vehicles/GTI_Left/*.png'
pathVehicles3 = 'data/vehicles/GTI_MiddleClose/*.png'
pathVehicles4 = 'data/vehicles/GTI_Right/*.png'
pathVehicles5 = 'data/vehicles/KITTI_extracted/*.png'

NonVehicles_path = [pathNonVehicles1, pathNonVehicles2]
Vehicles_path = [pathVehicles1, pathVehicles2, pathVehicles3, pathVehicles4, pathVehicles5]
NonVehicles = []
Vehicles = []

# Cteate list of image directories for Vehicles and non Vehicles
for path in Vehicles_path:
    Vehicles.extend(glob.glob(path))

for path in NonVehicles_path:
    NonVehicles.extend(glob.glob(path))
```

```python
# Load images
featureNonVehicles = []
featureVehicles = []

for img_path in NonVehicles:
    img = cv2.imread(img_path)
    featureNonVehicles.append(img)

for img_path in Vehicles:
    img = cv2.imread(img_path)
    featureVehicles.append(img)
```

## In the next cells I tested different combinations of color space, histogram bins and hof filter

### Test of Color Histogram accuracy performance
Here I tested different image color space representations: 'HSV', 'LUV', 'HLS', 'YUV', 'YCrCb', 'LAB'
And tested SVM on each of them to choose the best one.
Standard scallar was applyed before SVM

```python
cSpace = ['HSV', 'LUV', 'HLS', 'YUV', 'YCrCb', 'LAB']
hisBins = [4, 8, 16, 32, 64]

temp_V = []
temp_nonV = []
results_hist = []
# Check the training time for the SVC
t=time.time()
for col in cSpace:
    col_transformation = eval('cv2.COLOR_BGR2' + col)
    temp_V = [cv2.cvtColor(i, col_transformation) for i in featureVehicles]
    temp_nonV = [cv2.cvtColor(i, col_transformation) for i in featureNonVehicles]
    # Define the labels vector
    y = np.hstack((np.zeros(len(temp_nonV)), np.ones(len(temp_V))))
    for hisB in hisBins:
        fVehicles = [color_hist(i, hisB) for i in temp_V]
        fNonVehicles = [color_hist(i, hisB) for i in temp_nonV]
        X = np.vstack((fNonVehicles, fVehicles)).astype(np.float64)
        # Fit a per-column scaler
        X_scaler = StandardScaler().fit(X)
        # Apply the scaler to X
        scaled_X = X_scaler.transform(X)
        # Split up data into randomized training and test sets
        rand_state = 42 #np.random.randint(0, 100)
        X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.2, random_state=rand_state)
        # Use a linear SVC 
        svc = LinearSVC()
        svc.fit(X_train, y_train)
        # Check the score of the SVC
        rez = {'ColorSpaca': col, 'HistBin': hisB, 'Accuracy': round(svc.score(X_test, y_test), 4)}
        results_hist.append(rez)
        print(rez)
t2 = time.time()
print(round(t2-t, 2), 'Seconds to train...')
```

#### The best 5 results

```python
pd.DataFrame(results_hist).sort_values(by='Accuracy', ascending=False).head(5)
```

### Build classifier on features combinations (histogram, spatial and HOG)

```python
Features_Hist = [{'HcolorSpace': 'HSV', 'hBin': 64},
                 {'HcolorSpace': 'LAB', 'hBin': 64},
                 {'HcolorSpace': 'LAB', 'hBin': 32},
                 {'HcolorSpace': 'LAB', 'hBin': 16}]

Features_Spat = [{'ScolorSpace': 'LAB', 'sBin': 16},
                 {'ScolorSpace': 'LAB', 'sBin': 12}]

Features_HOG  = [{'colorSpace': 'LAB',   'Orient': 10, 'Pix': 16, 'Channel':1, 'Block': 4},
                 {'colorSpace': 'YCrCb', 'Orient': 6,  'Pix': 8,  'Channel':1, 'Block': 8},
                 {'colorSpace': 'YCrCb', 'Orient': 6,  'Pix': 16, 'Channel':1, 'Block': 4},
                 {'colorSpace': 'LUV',   'Orient': 10, 'Pix': 16, 'Channel':1, 'Block': 4}]

# the 4th element
# {'colorSpace': 'LUV',   'Orient': 10, 'Pix': 16, 'Channel':1, 'Block': 4}
```

```python
y = np.hstack((np.zeros(len(featureNonVehicles)), np.ones(len(featureVehicles))))
results_all = [] # to save all results
rez = {} # to save temporary results
# Histogram
t1 = time.time()
for h in Features_Hist:
    # Color transformation
    h_color = eval('cv2.COLOR_BGR2' + h['HcolorSpace'])
    h_bins = h['hBin']
    h_V =    [cv2.cvtColor(i, h_color) for i in featureVehicles]
    h_nonV = [cv2.cvtColor(i, h_color) for i in featureNonVehicles]
    # Histigram features
    h_feat_V =    [color_hist(i, h_bins) for i in h_V]
    h_feat_nonV = [color_hist(i, h_bins) for i in h_nonV]
    # Spatial
    for s in Features_Spat:
        # Color transformation
        s_color = eval('cv2.COLOR_BGR2' + s['ScolorSpace'])
        s_bins = s['sBin']
        s_V =    [cv2.cvtColor(i, s_color) for i in featureVehicles]
        s_nonV = [cv2.cvtColor(i, s_color) for i in featureNonVehicles]
        # Spatial features
        s_feat_V =    [bin_spatial(i, (s_bins,s_bins)) for i in s_V]
        s_feat_nonV = [bin_spatial(i, (s_bins,s_bins)) for i in s_nonV]
        #HOG
        for ho in Features_HOG:
            # Color transformation
            ho_color = eval('cv2.COLOR_BGR2' + ho['colorSpace'])
            ho_orient = ho['Orient']
            ho_pix = ho['Pix']
            ho_channel = ho['Channel']
            ho_block = ho['Block']
            ho_V =    [cv2.cvtColor(i, ho_color) for i in featureVehicles]
            ho_nonV = [cv2.cvtColor(i, ho_color) for i in featureNonVehicles]
            # HOG features
            ho_feat_V =    [get_hog_features(i[:,:,ho_channel], ho_orient, ho_pix, ho_block,
                                             vis=False, feature_vec=True) for i in ho_V]
            ho_feat_nonV = [get_hog_features(i[:,:,ho_channel], ho_orient, ho_pix, ho_block,
                                             vis=False, feature_vec=True) for i in ho_nonV]
            t_c_1 = time.time()
            # Classification 1
            # Features concatenation
            histIsUsed, spatialIsUsed = True, True
            X = np.vstack((np.hstack((h_feat_nonV, s_feat_nonV, ho_feat_nonV)),
                           np.hstack((h_feat_V,    s_feat_V,    ho_feat_V))))
            acc = SVM_result(X, y)
            rez = {}
            rez.update(h)
            rez.update(s)
            rez.update(ho)
            rez['Accuracy'] = acc
            rez['histIsUsed'] = histIsUsed
            rez['spatialIsUsed'] = spatialIsUsed
            results_all.append(rez.copy())
            print(rez)
            # Classification 2
            # Features concatenation
            histIsUsed, spatialIsUsed = True, False
            X = np.vstack((np.hstack((h_feat_nonV, ho_feat_nonV)),
                           np.hstack((h_feat_V,    ho_feat_V))))
            # Fit a per-column scaler
            acc = SVM_result(X, y)
            rez['Accuracy'] = acc
            rez['histIsUsed'] = histIsUsed
            rez['spatialIsUsed'] = spatialIsUsed
            results_all.append(rez.copy())
            print(rez)
            # Classification 3
            # Features concatenation
            histIsUsed, spatialIsUsed = False, True
            X = np.vstack((np.hstack((s_feat_nonV, ho_feat_nonV)),
                           np.hstack((s_feat_V,    ho_feat_V))))
            # Fit a per-column scaler
            acc = SVM_result(X, y)
            rez['Accuracy'] = acc
            rez['histIsUsed'] = histIsUsed
            rez['spatialIsUsed'] = spatialIsUsed
            results_all.append(rez.copy())
            print(rez)
            t_c_2 = time.time()
            print(round(t_c_2-t_c_1, 2), 'Seconds to train 3 types of classificators')
t2 = time.time()
print(round(t2-t1, 2), 'Seconds to train all classificators')
```

### The final version of classifier

```python
y = np.hstack((np.zeros(len(featureNonVehicles)), np.ones(len(featureVehicles))))

ho = Features_HOG[3]
h =  Features_Hist[1]
s =  Features_Spat[1]

# Feature parameters:
h_color = eval('cv2.COLOR_BGR2' + h['HcolorSpace'])
h_bins = h['hBin']

s_color = eval('cv2.COLOR_BGR2' + s['ScolorSpace'])
s_bins = s['sBin']

ho_color = eval('cv2.COLOR_BGR2' + ho['colorSpace'])
ho_orient = ho['Orient']
ho_pix = ho['Pix']
ho_channel = ho['Channel']
ho_block = ho['Block']

# Color transformation
h_V =     [cv2.cvtColor(i, h_color)  for i in featureVehicles]
h_nonV =  [cv2.cvtColor(i, h_color)  for i in featureNonVehicles]

s_V =     [cv2.cvtColor(i, s_color)  for i in featureVehicles]
s_nonV =  [cv2.cvtColor(i, s_color)  for i in featureNonVehicles]

ho_V =    [cv2.cvtColor(i, ho_color) for i in featureVehicles]
ho_nonV = [cv2.cvtColor(i, ho_color) for i in featureNonVehicles]

# Features
h_feat_V =    [color_hist(i, h_bins) for i in h_V]
h_feat_nonV = [color_hist(i, h_bins) for i in h_nonV]

s_feat_V =    [bin_spatial(i, (s_bins,s_bins)) for i in s_V]
s_feat_nonV = [bin_spatial(i, (s_bins,s_bins)) for i in s_nonV]

ho_feat_V =   [get_hog_features(i[:,:,ho_channel],ho_orient,ho_pix,ho_block,vis=False,feature_vec=True) for i in ho_V]
ho_feat_nonV =[get_hog_features(i[:,:,ho_channel],ho_orient,ho_pix,ho_block,vis=False,feature_vec=True) for i in ho_nonV]

# Classification
X = np.vstack((np.hstack((h_feat_nonV,
                          s_feat_nonV,
                          ho_feat_nonV)),
               np.hstack((h_feat_V,
                          s_feat_V,
                          ho_feat_V))))

Accuracy, clf, X_scaler = SVM_result(X, y, model_output = True)
print("Accuracy of classifier: {}%".format(100*Accuracy))
```

### Image processing for video

```python
ho = Features_HOG[3]
h =  Features_Hist[1]
s =  Features_Spat[1]

# Feature parameters:
#h_color = eval(color_from + h['HcolorSpace'])
h_bins = h['hBin']

#s_color = eval(color_from + s['ScolorSpace'])
s_bins = s['sBin']

#ho_color = eval(color_from + ho['colorSpace'])
ho_orient = ho['Orient']
ho_pix = ho['Pix']
ho_channel = ho['Channel']
ho_block = ho['Block']

def color_space(h, s, ho):
    color_from = 'cv2.COLOR_BGR2'
    return eval(color_from + h), eval(color_from + s), eval(color_from + ho)

def car_detection(img, clf, X_scaler, img_examples = False):
    # Color transformation
    h_color, s_color, ho_color = color_space(h['HcolorSpace'], s['ScolorSpace'], ho['colorSpace'])
    h_img  = cv2.cvtColor(img, h_color)
    s_img  = cv2.cvtColor(img, s_color)
    ho_img = cv2.cvtColor(img, ho_color)
    # Features
    h_feat  = color_hist(h_img, h_bins)
    s_feat  = bin_spatial(s_img, (s_bins,s_bins))
    ho_feat = get_hog_features(
        ho_img[:,:,ho_channel],
        ho_orient,
        ho_pix,
        ho_block,
        vis=False,
        feature_vec=True)
    X = np.hstack((h_feat, s_feat, ho_feat))
    X = X_scaler.transform(X.reshape(1, -1))
    if not img_examples:
        return clf.predict(X)
    else:
        _, HOG_IMG = get_hog_features(
            ho_img[:,:,ho_channel],
            ho_orient,
            ho_pix,
            ho_block,
            vis=True,
            feature_vec=True)
        return clf.predict(X), HOG_IMG
```

### HOG Sub-sampling Window Search

```python
ho = Features_HOG[3]
h =  Features_Hist[1]
s =  Features_Spat[1]

# Function that can extract features using hog sub-sampling and make predictions
def find_cars(img, ystart, ystop, scale, svc, X_scaler, orient, pix_per_cell, cell_per_block, spatial_size,
              hist_bins, ho_channel, cells_per_step = 1, window = 64, Boxlist = False):
    # Window = 64 was the orginal sampling rate
    # cells_per_step = 1 Instead of overlap, define how many cells to step
    
    # Color transformation
    img_tosearch = img[ystart:ystop,:,:]
    
    if scale != 1:
        imshape = img_tosearch.shape
        img_tosearch = cv2.resize(img_tosearch, (np.int(imshape[1]/scale), np.int(imshape[0]/scale)))
    
    imshape = img_tosearch.shape
    h_color, s_color, ho_color = color_space(h['HcolorSpace'], s['ScolorSpace'], ho['colorSpace'])
    ho_img = cv2.cvtColor(img_tosearch, ho_color)
    
    nxblocks = (imshape[1] // pix_per_cell) - cell_per_block + 1
    nyblocks = (imshape[0] // pix_per_cell) - cell_per_block + 1 
    nfeat_per_block = orient*cell_per_block**2
    
    nblocks_per_window = (window // pix_per_cell) - cell_per_block + 1
    
    nxsteps = (nxblocks - nblocks_per_window) // cells_per_step
    nysteps = (nyblocks - nblocks_per_window) // cells_per_step
    
    # Individual channel HOG features for the entire image
    hog = get_hog_features(ho_img[:,:,ho_channel], orient, pix_per_cell, cell_per_block,vis=False, feature_vec=False)
    # To collect all box to draw rectangles
    bbox_list = []
    for xb in range(nxsteps):
        for yb in range(nysteps):
            ypos = yb*cells_per_step
            xpos = xb*cells_per_step
            # Extract HOG for this patch
            hog_features = hog[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 

            xleft = xpos*pix_per_cell
            ytop = ypos*pix_per_cell

            # Extract the image patch
            subimg = cv2.resize(img_tosearch[ytop:ytop+window, xleft:xleft+window], (64,64))
            
            # Get color features
            spatial_features  = bin_spatial(cv2.cvtColor(subimg, s_color), (spatial_size, spatial_size))
            hist_features = color_hist(cv2.cvtColor(subimg, h_color), hist_bins)

            # Scale features and make a prediction
            test_features = X_scaler.transform(np.hstack((hist_features, spatial_features, hog_features)).reshape(1, -1))    
            test_prediction = svc.predict(test_features)
            if test_prediction == 1:
                xbox_left = np.int(xleft*scale)
                ytop_draw = np.int(ytop*scale)
                win_draw = np.int(window*scale)
                bbox_list.append(((xbox_left, ytop_draw+ystart),(xbox_left+win_draw,ytop_draw+win_draw+ystart)))
                #cv2.rectangle(draw_img,(xbox_left, ytop_draw+ystart),(xbox_left+win_draw,ytop_draw+win_draw+ystart),(0,0,255),6) 
    if Boxlist:
        return bbox_list
    else:
        return draw_boxes(img, bbox_list)
```

### Car detection

```python
y1 = 360
y2 = 648
sc = 1.5

h_bins = h['hBin']
s_bins = s['sBin']
ho_orient = ho['Orient']
ho_pix = ho['Pix']
ho_channel = ho['Channel']
ho_block = ho['Block']

def find_cars_light(img, Boxlist_ = False):
    rez = find_cars(img = img, ystart = y1, ystop = y2, scale = sc, svc = clf, X_scaler = X_scaler,
                    orient = ho_orient, pix_per_cell = ho_pix, cell_per_block = ho_block,
                    spatial_size = s_bins, hist_bins = h_bins, ho_channel = ho_channel,
                    cells_per_step = 1, window = 64, Boxlist = Boxlist_)
    if Boxlist_:
        return rez
    else:
        return cv2.cvtColor(rez, cv2.COLOR_BGR2RGB)
```

### Heatmap

```python
def add_heat(bbox_list, shape_ = (720, 1280)):
    # Iterate through list of bboxes
    heatmap = np.zeros(shape=shape_, dtype=np.float)
    for box in bbox_list:
        # Add += 1 for all pixels inside each bbox
        # Assuming each "box" takes the form ((x1, y1), (x2, y2))
        heatmap[box[0][1]:box[1][1], box[0][0]:box[1][0]] += 1
    # Return updated heatmap
    return heatmap# Iterate through list of bboxes

def threshold(heatmap, threshold):
    # Zero out pixels below the threshold
    heatmap[heatmap <= threshold] = 0
    # Return thresholded map
    return np.clip(heatmap, 0, 255)

def draw_labeled_bboxes(img, labels):
    # Iterate through all detected cars
    for car_number in range(1, labels[1]+1):
        # Find pixels with each car_number label value
        nonzero = (labels[0] == car_number).nonzero()
        # Identify x and y values of those pixels
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        # Define a bounding box based on min/max x and y
        bbox = ((np.min(nonzerox), np.min(nonzeroy)), (np.max(nonzerox), np.max(nonzeroy)))
        # Draw the box on the image
        cv2.rectangle(img, bbox[0], bbox[1], (0,0,255), 6)
    # Return the image
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
%matplotlib inline
f, ax = plt.subplots(6, 3, figsize=(24,32))
for i in range(6):
    test_img = cv2.imread('test_images/test' + str(i+1) + '.jpg')
    img_all_boxes = find_cars_light(test_img)
    imgBoxlist = find_cars_light(img= test_img, Boxlist_ = True)
    img_heat = threshold(add_heat(imgBoxlist), 1)
    img_heat_boxes = draw_labeled_bboxes(np.copy(test_img), label(img_heat))
    
    ax[i,0].imshow(img_all_boxes)
    ax[i,0].set_title('Image with all boxes', fontsize=10)
    ax[i,1].imshow(img_heat, cmap = 'hot')
    ax[i,1].set_title('Heat map', fontsize=10)
    ax[i,2].imshow(img_heat_boxes)
    ax[i,2].set_title('Image after threshold', fontsize=10);
```

```python
heatmap_ = []
def pipline_(img):
    global heatmap_
    img_color = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    imgBoxlist = find_cars_light(img = img_color, Boxlist_ = True)
    heat = add_heat(imgBoxlist)
    if len(heatmap_) >= 7:
        heatmap_.pop(0)
        heatmap_.append(heat)
    else:
        heatmap_.append(heat)
    img_heat = threshold(np.mean(heatmap_,axis=0), 3)
    return draw_labeled_bboxes(img_color, label(img_heat))
```
